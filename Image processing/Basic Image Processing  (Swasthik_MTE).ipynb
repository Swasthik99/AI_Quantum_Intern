# Accessing and Modifying pixel values
 You can access a pixel value by its row and column coordinates For BGR image, it returns an array of Blue, Green, Red
values.For grayscale image, just corresponding intensity is returned

```python
import cv2
import matplotlib.pyplot as plt
#Accessing and Modifying pixel values
bgr_img = cv2.imread( 'leana.jpeg')
# Reading BLue-Green-Red image
plt.imshow(bgr_img)
plt.title("input color image")

plt.show()
#You can access a pixel value by its row and column coordinates.
# For BGR image, it returns an array of Blue, Green, Red values.
# For grayscale image, just corresponding intensity is returned.
px = bgr_img[150,150]
print(px)

# Accessing only blue pixel value
blue = bgr_img[100,100,0]
print(blue)

# Accessing only gree pixel value
green = bgr_img[150,150,1]
print(green)
# Accessing only red pixel value
red = bgr_img[100,100,0]
print(red)



```

## Basic Image Conversion



```python
#convert a color image to greyscale image without using any Library function

import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
img = image.imread('A122.tif')
#Reading RGB image

b=img[:,:,0]
g=img[:,:,1]
r=img[:,:,2]
plt.imshow(img)
plt.title("Input BGR color image")
# to hide tick values on X and Y axis plt.show()
#Conversion from Color to grayscale
prop=img.shape

for i in range(prop[0]):
    for j in range( prop[1]):
       
         img[i][j]=(r[i][j]*0.2989)+ (g[i][j]*0.5870)+ (b[i][j]*0.114)
        
cv2.imwrite("leana.jpeg",img)
plt.imshow(img)
plt.title("Greyscale image")
# to hide tick values on X and Y axis
plt.show()
```

# Comparison of Processing time of different forms of implemenattion


```python
#convert a color image to greyscale image without using any Library function

import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib. image as image


img= cv2.imread('opencv-logo.jfif')
b=img[:,:,0]
g=img[:,:,1]
r=img[:,:,2]

plt.imshow(img)

plt.title('Input BGR color image')

# to hide tick values on X and Y axis

plt.show()

#Conversion from Color to grayscale
#Method-1 scratch method
prop = img.shape
                
import time
t10 = time.time()
for i in range(prop[0]):
    for j in range( prop[1]):

        img[i][j]=(r[i][j]*0.2989)+ (g[i][j]*0.5870)+(b[i][j]*0.114)

t11 = time.time()
time_forloop = t11 - t10

print(time_forloop)

cv2.imwrite('leana.jpeg', img)

plt.imshow(img)

plt.title('Greyscale image')

# to hide tick values on X and Y axis

plt.show()

#Method-2: Using CV2 function cv2.COLOR_BGR2GRAY

t20 = time.time()
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
t21 = time.time()
time_cv2 = t21-t20
print(time_cv2)
plt.imshow(gray_image, cmap='gray')
plt.title('COLOR_BGR2GRAY')

plt.show()

#method-3: Simpe mean method
t30 = time.time()
img2 =np.mean(img, axis=2)
t31 = time.time()
time_mean = t31 - t30
print(time_mean)
print(img2.shape)
print(img.shape)
plt.imshow(img2,'gray')
plt.title('Mean_grayscale image')

plt.show()

#Comparision of Processing Speed  of above three diffrent forms of implementation
x= ['Method-1', 'Method-2', 'Method-3']
Processing_Times = [time_forloop, time_cv2, time_mean]

# x_pos[i for i, in enumerate(x)] 
x_pos = [i for i in range(len(x))]

#x_pos[i for i, in enumerate(x)]

plt.bar(x_pos, Processing_Times, color="blue")
plt.xlabel("<----------Method----------->")
plt.ylabel("<-------Processing Time------>")
plt.title("Processing Time Vs Method ")

plt.xticks(x_pos, x)

plt.show()
```

# Exercise: Covert Color to grayscale image using def function and list of comprehension method

```python
#convert an image to binary image
import cv2 
import matplotlib.pyplot as plt

#open the image in greyscale

#img=cv2.imread("Lenna.tif", 0) 
img=cv2.imread('opencv-logo.jfif',0) 

plt.imshow(img, 'gray')
plt.title('greyscale image')


plt.show()

#if the intensity is lesser than equal to 127,make it black else white
prop=img.shape

for i in range(prop[0]):
    for j in range( prop[1]):
        if img[i][j] <= 127:
            img[i][j] = 0
        else:
            img[i][j] = 255

cv2.imwrite('leana.jpeg',img)
img = cv2.imread('leana.jpeg')
plt.imshow(img, "gray")
plt.title('binary image')

#plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis

plt.show()

# Using CV2 function

gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(blackAndWhiteImage, 'gray')
plt.title('blackAndWhiteImage')
plt.show()
```

```python
# organizing imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image

#path to input images are specified and images are loaded with imread commond
img =cv2.imread('img11.jfif')
image1 = cv2.resize(img, (512, 512))
img1 = cv2.imread('img12.jfif')
image2 = cv2.resize(img1, (512, 512))
plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()

#cv2.addWeighted is applied over the image

#image inputs with applied parameters
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.3, 0)
plt.imshow(weightedSum)
plt.show()

#path to input images are specified and images are loaded with imread commond
#image1 = cv2.imread('input1.jfif')
#image2 = cv2.imread('input2.jfif')
plt.imshow(image1)
plt.show()
#with the subtracted image
plt.imshow(image2)
plt.show()

#cv2.substract is applied over the image inputs with applied parameters
add = cv2.add(image1, image2)
sub = cv2.subtract(image1, image2)
mul = cv2.multiply(image1, image2)

#the window showing output image with the addition image
plt.imshow(add)
plt.show()
#with the subtracted image
plt.imshow(sub)
plt.show()
#with the multiplicated image
plt.imshow(mul)
plt.show()


```
